```python
data = pd.read_csv('C:\\Users\\DHAVAL\\AppData\\Local\\Programs\\Python\\Python39\\all_data.csv')
data.head(10)
```

```python
data = data.dropna(how='all')
data['month']=data['Order Date'].str[0:2]
data = data[data['month']!='Or']
data['month'] = data['month'].astype('int32')
```

```python
data['sales_ttl'] = data['Quantity Ordered'].astype('int32')*data['Price Each'].astype('float')#this is useful only here cz types are still str in og data
# converting whole column into numeric is preferable method
data['Quantity Ordered'] = pd.to_numeric(data['Quantity Ordered'])# this will make it numeric for future prps
data['Price Each'] = pd.to_numeric(data['Price Each'])
data['month'] = pd.to_numeric(data['month'])

data.head(4)
```

```python
data.groupby('month').sum().sort_values(['sales_ttl'], ascending=False).head(12)
#show you highest to lowest sales monthwise
```

```python
#finding city with highest num of sales
data['city'] = data['Purchase Address'].apply(lambda x: x.split(',')[1]+x.split(',')[2] )
data.head(5)
```

```python
data.groupby('city').sum().sort_values(['sales_ttl'], ascending=False)
```
